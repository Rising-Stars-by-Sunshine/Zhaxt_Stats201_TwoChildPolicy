```python
import csv
import matplotlib.pyplot as plt
```

```python
file = csv.reader(open("002570.SZ.csv"))
next(file)
data = []
for line in file:
    try:
        data.append(float(line[4]))
    except:
        pass
n = len(data)
plt.scatter(range(n), data, s=1)
plt.show()
```

```python
before1 = 1107
before2 = 1314
n1 = before2 - before1
before = data[before1:before2]
plt.scatter(range(n1), before, s=1)
plt.show()

with open("before.csv", 'w', encoding = "utf-8", newline = '\n') as d:
    d.writelines('before')
    for value in before:
        d.writelines('\n' + str(value))
    d.close()
```

```python
after1 = 1314
after2 = 1514
n2 = after2 - after1
after = data[after1:after2]
plt.scatter(range(n2), after, s=1)
plt.show()

with open("after.csv", 'w', encoding = "utf-8", newline = '\n') as d:
    d.writelines('after')
    for value in after:
        d.writelines('\n' + str(value))
    d.close()
```
