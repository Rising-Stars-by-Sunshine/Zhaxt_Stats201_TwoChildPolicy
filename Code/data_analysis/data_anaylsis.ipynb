```python
import csv
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
file1 = csv.reader(open("before.csv"))
next(file1)
before = []
for line in file1:
        before.append(float(line[0]))
n1 = len(before)

file2 = csv.reader(open("after.csv"))
next(file2)
after = []
for line in file2:
        after.append(float(line[0]))
n2 = len(after)
```

```python
X1 = np.array([[i] for i in range(n1)])
y1 = np.array(before)

model1 = LinearRegression()
model1.fit(X1, y1)
predictions1 = model1.predict(X1)

plt.scatter(X1, y1, color='blue', s=1)
plt.plot(X1, predictions1, color='red')
plt.show()

k1 = model1.coef_[0]
b1 = model1.intercept_
print(k1, b1)
```

```python
X2 = np.array([[i] for i in range(n2)])
y2 = np.array(after)

model2 = LinearRegression()
model2.fit(X2, y2)
predictions2 = model2.predict(X2)

plt.scatter(X2, y2, color='blue', s=1)
plt.plot(X2, predictions2, color='red')
plt.show()

k2 = model2.coef_[0]
b2 = model2.intercept_
print(k2, b2)
```

```python
X = np.array([[i] for i in range(n1 + n2)])
plt.scatter(X[:n1], y1, color='blue', s=1)
plt.plot(X[:n1], predictions1, color='red')
plt.scatter(X[n1:], y2, color='purple', s=1)
plt.plot(X[n1:], predictions2, color='orange')
plt.show()

print("k:", k1, k2)
print("b:", b1, b2)
```
